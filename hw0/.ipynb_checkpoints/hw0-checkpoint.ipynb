## [Plotting normal distributed points, 5 points]

Generate $1000$ points from 2D multivariate normal
distribution having mean $\mu = \left[
\begin{array}{c}
  1  \\
  1
\end{array}
\right]$ and covariance $\Sigma =
\left[
  \begin{array}{rr}
    0.1 & -0.05 \\
    -0.05&  0.2
  \end{array}
\right]
$.  Define the function $f({\bf x}, r) := \frac{({\bf x} - \mu)^{ \top } * \Sigma^{-1} *
  ({\bf x} - \mu) }{ 2} - r $. On a single plot, show the following:
* The level sets $f({\bf x}, r) = 0$ for $r=1, 2, 3$.
* Scatter plot of randomly generated points with points lying
outside $f({\bf x} , 3) = 0$ showing in black while points inside shown in
blue.
* Title of the plot showing how many points lie outside $f({\bf
    x}, 3) = 0$.
Submit your final plot as well as your implementation.

```python
%matplotlib inline

import numpy as np
from numpy.linalg import inv
from numpy.random import multivariate_normal as mvnrnd
import matplotlib.pyplot as plt
import matplotlib

def f(x,mu,sigma,r):
    a = (x-mu).transpose() @ inv(sigma) @ (x-mu) / 2 - r
    return np.asscalar(a)

#matplotlib.rcParams['text.usetex'] = True
#matplotlib.rcParams['text.latex.unicode'] = True

mu = np.matrix([[1],[1]])
sigma = np.matrix([[0.1, -0.05],[-0.05, 0.2]])
a = np.vectorize(lambda x,y,r:f([[x],[y]],mu,sigma,r))
x = np.linspace(-1,3)
X,Y = np.meshgrid(x,x)

contour_colors = ('r','m','g')
plt.figure()
CS = plt.contour(X,Y,a(X,Y,3),[-2,-1,0],colors=contour_colors)
labels = ['$f(x,1)=0$', '$f(x,2)=0$','$f(x,3)=0$']
for label, collection in zip(labels, CS.collections):
    collection.set_label(label)

rnd_points = mvnrnd(np.squeeze(np.asarray(mu)),sigma,1000)
outside = a(rnd_points[:,0], rnd_points[:,1], 3) > 0
plt.plot(rnd_points[outside, 0], rnd_points[outside, 1],marker='.',color='black',linestyle='none',markersize=3, label='Outside')
plt.plot(rnd_points[~outside, 0], rnd_points[~outside, 1],marker='.',color='b',linestyle='none',markersize=3, label='Inside')
plt.axis([-0.5,2.5,-0.5,2.5]);
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('There are {} points outside the $f(x,3)=0$ contour line'.format(np.count_nonzero(outside)));
```

# Discussion

We note that a large majority of the sampled points are inside of the area depicted by the contour line. In fact there are 57 points outside of the contour line and 943 inside of it, which makes us expect this line to depict a 95% confidence interval for the distribution of the sampled points. Several simulations yield similar values. We want to investigate this result.

We let $Y = (\vec{x} - \mu)^T * \Sigma^{-1} * (\vec{x} - \mu)$. We know from statistics that $Y$ is $\chi^2$-distributed with two degrees of freedom, thus we can construct a 95% confidence interval by solving for $y$ in the following equation

$$
    P\left(Y > 2y\right) = 0.05,
$$

since we have that $Y/2 = r$ at the contour line. We find from the quantile function of $\chi^2$ for two degrees of freedom that $2y = 5.99$, which gives us $y = 2.995 \approx 3$, which corresponds to $r = 3$. Which agrees with our earlier observation that this contour line depicts a 95% confidence interval.

## [Covariance and correlation, 5 points]
Load dataset0.txt ($X$) containing 1074 data points
each having 12 features related to US schools. Compute the covariance
and correlation matrix for $X$. Scale each feature
in $X$ between $[0, 1]$ to obtain a new dataset $Y$. Compute the
covariance and correlation matrices for $X$ and $Y$, and plot them (e.g. as colormaps).
What do you observe? Show a scatter plot of the pair of features in $Y$ having minimum
correlation, indicating in the title the feature indices and the
correlation value. Submit the plots, comments and your implementation.

```python
def data_normalization(data):
    minValues = data.min(0)
    maxValues = data.max(0)
    interval = maxValues-minValues
    return ((data - minValues)/interval).transpose()
    
data = np.loadtxt("dataset0.txt")
normed_data = data_normalization(data)
data = data.transpose()
corr = np.corrcoef(normed_data)

def plotty(title, data):
    im = plt.matshow(data,cmap='seismic')
    plt.colorbar(im, fraction=0.046, pad=0.04)
    plt.title(title, y=1.1)
    
args = [('Raw data - Correlation', np.corrcoef(data)),
        ('Scaled data - Correlation', np.corrcoef(normed_data)),
        ('Raw data - Covariance', np.cov(data)),
        ('Scaled data - Covariance', np.cov(normed_data))]

for title, x in args:
    plotty(title, x)

ind1, ind0 = np.unravel_index(np.abs(corr).argmin(), corr.shape)
plt.figure()
plt.axis('equal')
plt.xlabel('Feature {}'.format(ind1))
plt.ylabel('Feature {}'.format(ind0))
plt.plot(normed_data[ind1,:],normed_data[ind0,:],'.',markersize=4)
plt.title('Correlation between feature {} and feature {} is {:.3f}'.format(ind0, ind1, corr[ind0,ind1]))
```

# Discussion

We see that the correlation matrix is independent of scaled data, which is expected since this is just the normalized covariance matrix. The covariance matrix of the scaled data is similar to the correlation matrices, except for a scaling factor. However, it differs significantly from the covariance matrix of the raw data, as the numerical values of the different features play an important role here. For example, the covariance of a parameter with itself is just the variance, and this is of course a high number when the numerical values are high, even if it's not very spread out.

The scatter plot shows the two features with the least correlation. We see this as there is obviously no good way to make a linear fit to the data points with low residuals.
